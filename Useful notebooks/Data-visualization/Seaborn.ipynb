## Seaborn 
Seaborn is a Python data visualization library based on Matplotlib. It provides a high-level interface for creating attractive graphs.

Command for Importing seaborn library<br>
<b>import seaborn as sns</b>

```python
import seaborn as sns
```

```python
#Importing pandas
import pandas as pd
# Load the Iris.csv dataset using read_csv() function of Pandas dataframe.
iris = pd.read_csv('iris.csv',names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
```

### Scatterplot
Function used : <br><b>sns.scatterplot(x='column_name',y='column_name',data='dataset_name')</b>

```python
sns.scatterplot(x='sepal_length', y='sepal_width', data=iris)
```

To hightlight the points by species, use the <b>hue</b> arguement.

```python
sns.scatterplot(x='sepal_length', y='sepal_width', hue='class', data=iris)
```

### Line Chart
Function Used : <br>
<b>sns.lineplot()</b> : The only arguement is the data which is Iris dataset<br>
<b>sns.kdeplot()</b> : Rounds the edges of the curves and makes the plot looks cleaner.

```python
sns.lineplot(data=iris.drop(['class'], axis=1))
```

### Histogram
Function Used: <br>
<b>sns.distplot()</b><br>
Pass the column you want to plot which is <b>'petal_length'</b> of Iris dataset in the below case.<br>
Number of <b>bins</b> and <b>Gaussian Kernel density estimate</b> can also be passed inside the graph<br>

```python
#Gaussian Kernel density estimate is set to False
sns.distplot(iris['petal_length'], bins=10, kde=False)
```

```python
#Gaussian Kernel density estimate is set to True
sns.distplot(iris['petal_length'], bins=10, kde=True)
```

### Bar Chart
Function used : <br>
<b>sns.countplot()</b>

```python
sns.countplot(iris['petal_width'])
```
